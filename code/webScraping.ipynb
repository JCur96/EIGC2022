```python
# combine all text from each web page? 
paragraphs = soup.find_all('p')
paragraphs2 = soup2.find_all('p')
```

```python
import pandas as pd
textDf = pd.DataFrame(columns=['paraIndex', 'text'])
for i in range(len(paragraphs)):
    text = paragraphs[i].get_text()
    # textDf.loc[len(textDf):,'paraIndex'] = i
    # textDf.loc[len(textDf):,'text'] = text
    textDf.loc[len(textDf)] = [i, text]

    
```

```python
# maybe want to collapse to "sinlge" article
listOftext = []
for i in range(len(paragraphs)):
    text = paragraphs[i].get_text()
    listOftext.append(text)

fullText = ''.join(listOftext)


listOftext2 = []
for i in range(len(paragraphs2)):
    text = paragraphs2[i].get_text()
    listOftext2.append(text)

alltext = listOftext + listOftext2
fullText = ''.join(alltext)

#print(fullText)
```

```python
import nltk
nltk.download()
```

```python
# now do keywords
from rake_nltk import Metric, Rake

r = Rake(ranking_metric=Metric.WORD_DEGREE, min_length=2, max_length=5) # Metric.WORD_DEGREE, and DEGREE_TO_FREQUENCY_RATIO and Metric.WORD_FREQUENCY

# Extraction given the text.
r.extract_keywords_from_text(fullText)

# Extraction given the list of strings where each string is a sentence.
# r.extract_keywords_from_sentences(<list of sentences>)

# To get keyword phrases ranked highest to lowest.
r.get_ranked_phrases()

# To get keyword phrases ranked highest to lowest with scores.
r.get_ranked_phrases_with_scores()

keywords = r.get_ranked_phrases_with_scores()
print("keywords: ", keywords)
```

```python
from keybert import KeyBERT
kw_model = KeyBERT()
keywords = kw_model.extract_keywords(fullText)

print(kw_model.extract_keywords(fullText, keyphrase_ngram_range=(3, 5), stop_words=None))
```
